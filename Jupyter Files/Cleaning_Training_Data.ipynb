```python
import pandas as pd
import numpy as np
```

```python
file_path1 = "C:/Users/hp/Documents/GitHub/Customer-Churn-ML-Prediction/Datasets/data_1.csv"
file_path2 = "C:/Users/hp/Documents/GitHub/Customer-Churn-ML-Prediction/Datasets/data_3.csv"

data1 = pd.read_csv(file_path1)
data2 = pd.read_csv(file_path2)
```

```python
# Merging the 2 training datasets
main_df = pd.concat([data1,data2])
```

# Cleaning Training data

```python
# Checking for empty cells
main_df.isnull().any()
```

## Handling "Column1"

```python
main_df.Column1
```

```python
# Removing Column1 from the dataset
main_df.drop('Column1', inplace=True, axis=1)
```

```python
main_df
```

```python
# Checking for empty cells
main_df.isnull().any()
```

## Replacing values in multiple columns

```python
# Replace values in multiple columns
columns_to_replace = ['MultipleLines', 'PhoneService', 'DeviceProtection', 'InternetService', 
                      'OnlineSecurity', 'OnlineBackup', 'StreamingTV', 'TechSupport', 'StreamingMovies', 
                      'Churn', 'SeniorCitizen', 'Partner', 'Dependents', 'PaperlessBilling']

# Define the mapping of values to replace
value_mapping = {
    True: 'Yes',
    False: 'No',
    np.nan: 'No',
    '': 'No',
    'NaN': 'No',
    'Nan': 'No',
    'NaN': 'No',
    0: 'No',
    1: 'Yes',
    'No phone service': 'No',
    'No internet service': 'No'
}

# Loop through each column and replace the values
for column in columns_to_replace:
    main_df[column] = main_df[column].replace(value_mapping)
```

## Handling column "TotalCharges"

```python
main_df["TotalCharges"].isnull().sum()
```

```python
a = main_df["TotalCharges"].dropna() # dropping null values 
```

```python
a = a.astype(str) # changing to string to replace non int values
```

```python
# replacing values

a = a.apply(lambda x : x.replace('',"0"))
a = a.apply(lambda x : x.replace("0 0","0"))
```

```python
a = a.astype(float) # changing to float to calculate mean
```

```python
my_mean = a.mean()
```

```python
main_df["TotalCharges"] = main_df["TotalCharges"].fillna(my_mean) # filling null values by mean
```

```python
main_df["TotalCharges"].isnull().sum()
```

```python
main_df["TotalCharges"] = main_df["TotalCharges"].apply(lambda x : str(x).replace(" ","0"))
```

```python
main_df["TotalCharges"] = main_df["TotalCharges"].astype(float)
```
