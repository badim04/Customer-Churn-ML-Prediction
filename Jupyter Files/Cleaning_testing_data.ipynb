```python
# Replace values in multiple columns
columns_to_replace = ['MultipleLines', 'PhoneService', 'DeviceProtection', 'InternetService', 
                      'OnlineSecurity', 'OnlineBackup', 'StreamingTV', 'TechSupport', 'StreamingMovies',
                      'SeniorCitizen', 'Partner', 'Dependents', 'PaperlessBilling']

# Define the mapping of values to replace
value_mapping = {
    np.nan: 'No',
    '': 'No',
    'NaN': 'No',
    'Nan': 'No',
    'NaN': 'No',
    0: 'No',
    1: 'Yes',
    'No phone service': 'No',
    'No internet service': 'No'
}

# Loop through each column and replace the values
for column in columns_to_replace:
    df[column] = df[column].replace(value_mapping)
```

## Handling column "TotalCharges"

```python
# Checking the unique values of the column
df["TotalCharges"].unique()
```

```python
# changing to string to replace non int values
df.TotalCharges = df.TotalCharges.astype(str)
```

```python
# replacing values
df.TotalCharges = df.TotalCharges.apply(lambda x: x.replace('', "0"))
df.TotalCharges = df.TotalCharges.apply(lambda x: x.replace("0.0", "0"))
df.TotalCharges = df.TotalCharges.apply(lambda x: x.replace("000 000", "0"))
df.TotalCharges = df.TotalCharges.apply(lambda x: x.replace('0000000 0000000', "0"))
df.TotalCharges = df.TotalCharges.apply(lambda x: x.replace('0 0', '0'))
```

```python
# Changing to float
df.TotalCharges = df.TotalCharges.astype(float)
```

```python
#Getting the mean
df.TotalCharges.mean()
```
